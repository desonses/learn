```
import numpy as np
import matplotlib.pyplot as plt

```

Metodo numerico para el calculo del area bajo una curva, para este ejemplo de aplicacion es el metodo de los trapecios.

```
def funcion(x): # devuelve la función a evaluar
   # fx = np.sqrt(x)*np.sin(x)
    fx = np.sin(x)
    return fx

# regresa el area calculada, entre mayor numero de trapecios mayor presicion
# del calculo
def calcula_area(a, b, n): # calcula el area de la funcion fx usando trapecios
    h = (b - a) / n
    x = a
    
    suma = funcion(x) 

    for i in range(0, n - 1, 1):
        x += h
        suma += 2*funcion(x)
    suma += funcion(b)
    area = h*(suma / 2)
              
    return area
```

```

# intervalo de la funcion
a = 1
b = 3

n = 50 # numero de particiones

area = calcula_area(a, b, n)
print("Numero de intervalos: ", n)
print ("El area bajo la curva es: ", area)                  
```

```
muestras = n + 1
xi = np.linspace(a,b,muestras )
fi = funcion(xi)

# graficacion de la función

k = np.linspace(a, b, muestras * 10)
m = funcion(k)
plt.plot(k, m)

plt.fill_between(xi,0,fi,color='g')#rellena el area  
plt.title("Regla de los Trapecios")

for i in range(0, muestras, 1):#pinta los trapecios
    plt.axvline(xi[i], color ='w') 

plt.plot(xi,fi,'o')
plt.xlabel('x')
plt.ylabel('f(xi)')
plt.show()




```
