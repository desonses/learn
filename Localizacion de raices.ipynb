# Metodos numericos para el calculo de raices.

```

class Metodos:

    def __init__(self, func):
        self.func = func

    def f(self, valor):
        funcs = vars(math)
        libres = dict(x = valor)
        return (eval(self.func, funcs, libres))

    def derivada(self, x):
        return str(diff(self.func, x))

    def biseccion(self, a, b, n):
        """
        a : int, el valor del limite inferior

        b : int, el valor del limite superior

        n : int, numero de iteraciones a realizar

        """
        i = 1
        while i < n:
            c =  (a + b) / 2
            if self.f(c) == 0:
                print("sol exacta")
                break
        
            if self.f(c)*self.f(a) < 0:
                b = c
            if self.f(c)*self.f(a) > 0:
                a = c
            i = i + 1
        return c
    
    #################################################
    #################################################
    def newton(self, n , p0):
        """
        n : int, el numero de iteraiones a realizar

        p0 : float, el valor del punto inicial arbitrario
        """
        i = 1
        while i < n:
            try:
                p = p0 - (self.f(p0) / self.f(p0))
                #print(p)
                i+=1
                p0 = p
            except ValueError:
                print("Division por cero!")
                break
        return p

    #################################################
    #################################################
    def secante(self, a, b, n, e):
        """
        a: int, valor inferior del intervalo

        b:  int, valor superior del intervalo

        n: int, numero de iteraciones a realizar

        e: float, valor de aproximacion

        """
        
        i = 1
        while i <= n:
            p = b - (b - a ) * self.f(b) / (self.f(b) - self.f(a))
            if abs(p - b) < e:
                return p
            i = i + 1
            a = b
            b = p
    #############################################
    #############################################
    def falsa_posicion(self, a, b, e):
        """
        a : int, el valor del limite inferior

        b : int, el valor del limite superior

        e: float, el valor del error tolerado

        """

        xr = b - self.f(b)*(a - b) / (self.f(a) - self.f(b))
        valor = self.f(a) * self.f(b)

        if valor < 0:
            while abs(xr - a) > e:
                
                #print(xr)
                if self.f(a) * self.f(xr) == 0:
                    return xr
                if self.f(a) * self.f(xr) > 0:
                    a = xr
                if self.f(a) * self.f(xr) < 0:
                    b = xr
                xr = b - (self.f(b)*(a - b) / (self.f(a) - self.f(b)))
        else:
            print("Intervalo incorrecto o no esta definida la func")

        return xr

```

```
func = "(x)**3+4*(x)**2-1"
metodo = Metodos(func)

print("Metodo de Biseccion: ", metodo.biseccion(-1, 1, 15))
print("Metodo de Newton: ", metodo.newton(6, 0.5))
print("Metodo de Secante" ,metodo.secante(-1, 1, 20, 0.000001))
print("Metodo de Falsa posicion: ", metodo.falsa_posicion(0, 1, .002))

```
